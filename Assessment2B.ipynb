###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
file1 = r'C:\Users\gladi\OneDrive\Documents\guvi\task4\college_1.csv'
file2 = r'C:\Users\gladi\OneDrive\Documents\guvi\task4\college_2.csv'
ds = pd.read_csv(file1,sep = ',')
ds1 = pd.read_csv(file2,sep = ',')
display(ds)
display(ds1)
frame = [ds,ds1]
df = pd.concat(frame)
display(df)
```
